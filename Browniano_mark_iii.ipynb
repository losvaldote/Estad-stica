### Se generarán números aleatorios gaussianos y se guardarán en dos arreglos, uno para x y otro para y después vamos a obtener la trayectoria de una partícula con movimiento browniano en dos dimensiones. Puesto que la función que se va a usar genera número mayores que uno, se buscará el valor máximo generado y se reescalará en términos de este de tal forma que los número que se van a usar para calcular el desplazamiento cuadrático medio estén entre cero y uno.

```python
#Comencemos importando las librerías.

%matplotlib inline    

import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python
#Generamos 1000 números aleatorios gaussianos y se guardan en el arreglo X y en el arreglo y.
#Aquí podemos considerar como intervalo de tiempo desde 0 hasta 1 con 1000 pasos de tiempo.
X = np.random.randn(1000)
Y = np.random.randn(1000)
```

```python
#Vamos a hallar el máximo de cada arreglo para reescalarlo.
maxX = max(X)
maxY = max(Y)
```

```python
#Ahora que ya conocemos los valores máximos de cada arreglo procederemos a normalizar cada arreglo.
x = X/maxX
y = Y/maxY
```

### El desplazamiento cuadrático medio está dado por:

$$\left<\left(x(t) - x_0\right)^2\right>$$

### Donde $x_0$ es la posición inicial de la partícula.

### Desarrollando la expresión anterior, es posible reescribir el desplazamiento cuadrático medio como:

$$\left<(x(t))^2\right> + x_0^2 - 2x_0 \left<x(t)\right>$$

### La posición inicial para x es:

```python
x0 = x[0]
```

### La posición inicial para y es:

```python
y0 = y[0]
```

### Hagamos la gráfica de la trayectoria descrita por la partícula.

```python
plt.plot(x, y)
plt.title('Trayectoria')
plt.xlabel('x')
plt.ylabel('y')
```

### Vamos a calcular el promedio de $x^2$ y $y^2$.

```python
#Vamos a obtener un arreglo que guarde x^2 y otro que guarde y^2.
x2 = x**2
y2 = y**2
```

```python
#Vamos a calcular las medias para cada arreglo.
mediax2 = np.mean(x2)
mediay2 = np.mean(y2)
```

```python
mediay2
```

### Ahora vamos a calcular la media de x y y.

```python
mediax = np.mean(x)
mediay = np.mean(y)
```

### Ya que tenemos todo podemos calcular el desplazamiento cuadrático medio. Para x

```python
mediax2 + x[0]**2 - 2*X[0]*mediax
```

### Para y.

```python
mediay2 + y[0]**2 - 2*y[0]*mediay
```
